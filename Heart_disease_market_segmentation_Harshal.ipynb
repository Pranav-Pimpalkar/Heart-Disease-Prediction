```python
df1 = df.copy()
```

```python
# Remove the target column from the duplicate dataset df1
df1 = df1.drop(columns=['HeartDisease'])
```

```python
df1.head()
```

```python
df1.info()
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_features = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']

# Initialize LabelEncoders
label_encoders = {feature: LabelEncoder() for feature in categorical_features}

# Apply LabelEncoder to each categorical feature
for feature in categorical_features:
    df1[feature] = label_encoders[feature].fit_transform(df1[feature])
```

```python
from sklearn.preprocessing import StandardScaler
numeric_features = ['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak']
# Initialize StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to numerical features
df1[numeric_features] = scaler.fit_transform(df1[numeric_features])
```

```python
df1.head()
```

```python
from sklearn.cluster import KMeans
wcss  = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 10)
    kmeans.fit(df1)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11), wcss)

# set the axes and plot labels
# set the font size using 'fontsize'
plt.title('Elbow Plot', fontsize = 15)
plt.xlabel('No. of clusters (K)', fontsize = 15)
plt.ylabel('WCSS', fontsize = 15)

# display the plot
plt.show()
```

```python
cov_mat = np.cov(df1.T)
```

```python
eig_val, eig_vec = np.linalg.eig(cov_mat)
```

```python
eig_val = list(eig_val)

# 'sort(reverse = True)' will sort the eigenvalues in the descending order
eig_val.sort(reverse = True)

# print the sorted list
print(eig_val)
```

```python
percent_var = []

for i in eig_val:
    variation = (i/sum(eig_val))*100
    percent_var.append(variation)

percent_var
```

```python
np.cumsum(percent_var)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
```

```python
df2 = pca.fit_transform(df1)
df2.shape
```

```python
pca_df = pd.DataFrame(data = df2, columns = ['PCA1', 'PCA2','PCA3'])
pca_df.head()
```

```python
model = KMeans(3)
model.fit_predict(df2)
pca_df_kmeans = pd.concat([pca_df, pd.DataFrame({'cluster': model.labels_})], axis = 1)
```

```python
plt.figure(figsize = (10, 8))
sns.scatterplot(x = 'PCA1', y = 'PCA2', hue = 'cluster', data = pca_df_kmeans)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], marker = 'X', c = 'r', s = 80, label = 'centroids')
plt.title('Clustering using K-Means')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10, 8))
sns.scatterplot(x = 'PCA1', y = 'PCA3', hue = 'cluster', data = pca_df_kmeans)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], marker = 'X', c = 'r', s = 80, label = 'centroids')
plt.title('Clustering using K-Means')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10, 8))
sns.scatterplot(x = 'PCA2', y = 'PCA3', hue = 'cluster', data = pca_df_kmeans)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], marker = 'X', c = 'r', s = 80, label = 'centroids')
plt.title('Clustering using K-Means')
plt.legend()
plt.show()
```

```python
new_df = pd.concat([df1, pd.DataFrame({'Cluster': model.labels_})], axis = 1)
new_df.head()
```

```python
segment_description = new_df.groupby('Cluster').mean()
print("Segment Descriptions:\n", segment_description)

# Visualization
plt.figure(figsize=(12, 8))

# Plotting feature distributions by cluster
for feature in new_df.columns[:-1]:  # Excluding 'Cluster' column
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Cluster', y=feature, data=new_df)
    plt.title(f'Distribution of {feature} by Cluster')
    plt.show()
```

```python
segment_sizes = new_df['Cluster'].value_counts()
print("Segment Sizes:\n", segment_sizes)

# Display key characteristics of each segment
print("Segment Key Characteristics:\n", segment_description)
```

```python
target_segments = [0, 1]
df_target_segments = new_df[new_df['Cluster'].isin(target_segments)]

# Plot mean values for targeted segments
plt.figure(figsize=(12, 8))
for segment in target_segments:
    segment_data = df_target_segments[df_target_segments['Cluster'] == segment]
    mean_values = segment_data.drop(columns='Cluster').mean()
    plt.plot(mean_values, label=f'Segment {segment}')

plt.title('Feature Profiles of Target Segments')
plt.xlabel('Features')
plt.ylabel('Mean Value')
plt.legend()
plt.show()
```
